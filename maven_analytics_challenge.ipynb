### Importing the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Reading the csv file into a dataframe to view the data.

```python
marketing = pd.read_csv('marketing_data.csv')
marketing.head()
```

```python
marketing.describe()
```

### Looking for null values in the dataframe

```python
marketing.isnull().sum()
```

### As you can seen from the above data there are 24 null values for Income column and column as whitespaces

### Verifying the column name as you can see there is a space before and after income which can be difficult further

```python
marketing[' Income '].value_counts()
```

### Renamed the columns names of the dataframe to the existing one and removed the whitespaces using strip method for the column names

```python
marketing = marketing.rename(columns=lambda x: x.strip())
```

### As you can see from the below result there is no whitespaces for the income field but still there are 24 null-values.

```python
marketing.isnull().sum()
```

### Accessing the column name "Income" without spaces and "value_counts" counts number of occurences of each value in the field.

```python
marketing['Income'].value_counts()
```

### Looking deep into the values that contains null-values and trying to figure it out what we can do?

```python
marketing.loc[marketing['Income'].isnull()]
```

### Finding out the percentage of missing values in each column in the given dataset


```python
percent_missing = marketing.isnull().sum() * 100 / len(marketing)
missing_value_df = pd.DataFrame({'column_name': marketing.columns,
                                 'percent_missing': percent_missing})
missing_value_df
```

### As there are only 24 null-values which won't damage our analysis so dropped the null values from the dataframe.

```python
marketing = marketing.dropna()
```

### As you can see from the below result there are no more null-values in the dataframe.###

```python
marketing.isnull().sum()
```

### Creating a new csv file from the changes made to the old csv file

```python
marketing.to_csv('marketing_new_data.csv', index = False)
```
